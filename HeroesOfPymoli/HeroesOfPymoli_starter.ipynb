### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data['SN'].nunique()
df = pd.DataFrame({'Total Players': [total_players],})
df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_of_unique_items = purchase_data['Item ID'].nunique()
average_price = "${:.2f}".format(purchase_data['Price'].mean())
number_of_purchases = purchase_data['Purchase ID'].count()
total_revenue = "${:,.2f}".format(purchase_data['Price'].sum())

# Summary Dataframe
summary_df = pd.DataFrame({
                          'Number of Unique Items': [number_of_unique_items],
                          'Average Price': [average_price],
                          'Number of Purchases': [number_of_purchases],
                          'Total Revenue': [total_revenue],
                          })
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Group by "Gender"
gender_group_df = purchase_data.groupby("Gender")

# Count of gender group
gender_group_count_df = gender_group_df["SN"].nunique()

# Percentage by gender
gender_group_percent_df = ((gender_group_df["SN"].nunique()/total_players) * 100).map("{:.2f}".format)

# new Group by dataframe
gender_group_summary_df = pd.DataFrame({
                                        'Percentage of Players': gender_group_percent_df,
                                        'Total Count': gender_group_count_df,
                                      })

# Sort the values in descending order
gender_group_summary_df.sort_values(by=["Total Count"], ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Count of gender group
gender_group_count_df = gender_group_df["Purchase ID"].count()

# total purchase Value =
gender_group_total_purchase_df = gender_group_df["Price"].sum().map("${:,.2f}".format)

# Average Purchase price by gender
gender_group_avg_purchase_price_df = ((gender_group_df["Price"].sum()/gender_group_df["Purchase ID"].count())).map("${:.2f}".format)

# Average Purchase total per person
gender_group_avg_purchase_person_df = ((gender_group_df["Price"].sum()/gender_group_df["SN"].nunique())).map("${:.2f}".format)

# new Group by dataframe
gender_group_summary_df = pd.DataFrame({
                                        'Purchase Count': gender_group_count_df,
                                        'Average Purchase Price': gender_group_avg_purchase_price_df,
                                        'Total Purchase Value': gender_group_total_purchase_df,
                                        'Avg Purchase Total per Person': gender_group_avg_purchase_person_df,
                                      })
gender_group_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data['Age'], age_bins, labels=group_names)
age_group_df = purchase_data.groupby('Age Group')
age_group_count_df = age_group_df['Gender'].count()
age_group_percent_df = ((age_group_df['Gender'].count()/total_players) * 100).map("{:.2f}".format)
age_group_summary_df = pd.DataFrame({
                                    'Percentage of Players': age_group_percent_df,
                                    'Total Count': age_group_count_df,
                                    })
age_group_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Count of age group
age_group_count_df = age_group_df["Purchase ID"].count()

# total purchase Value =
age_group_total_purchase_df = age_group_df["Price"].sum().map("${:,.2f}".format)

# Average Purchase price by age
age_group_avg_purchase_price_df = ((age_group_df["Price"].sum()/age_group_df["Purchase ID"].count())).map("${:.2f}".format)

# Average Purchase total per person
age_group_avg_purchase_person_df = ((age_group_df["Price"].sum()/age_group_df["SN"].nunique())).map("${:.2f}".format)

# new Group by dataframe
age_group_summary_df = pd.DataFrame({
                                        'Purchase Count': age_group_count_df,
                                        'Average Purchase Price': age_group_avg_purchase_price_df,
                                        'Total Purchase Value': age_group_total_purchase_df,
                                        'Avg Purchase Total per Person': age_group_avg_purchase_person_df,
                                      })
age_group_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by SN
sn_group_df = purchase_data.groupby('SN')
sn_group_count_df = sn_group_df['SN'].count()
sn_group_total_purchase_df = sn_group_df['Price'].sum().map('${:.2f}'.format)
sn_group_avg_purchase_df = (sn_group_df['Price'].sum()/sn_group_df['Price'].count()).map('${:.2f}'.format)

sn_group_summary_df = pd.DataFrame({
                                    'Purchase Count': sn_group_count_df,
                                    'Average Purchase Price': sn_group_avg_purchase_df,
                                    'Total Purchase Value': sn_group_total_purchase_df,
                                    })
sn_group_summary_df.sort_values('Purchase Count', ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by SN
item_group_df = purchase_data.groupby(['Item ID', 'Item Name'])
item_group_count_df = item_group_df['Item ID'].count()
item_group_total_purchase_df = item_group_df['Price'].sum().map('${:.2f}'.format)
item_group_price_df = (item_group_df['Price'].sum()/item_group_df['Price'].count()).map('${:.2f}'.format)

item_group_summary_df = pd.DataFrame({
                                    'Purchase Count': item_group_count_df,
                                    'Item Price': item_group_price_df,
                                    'Total Purchase Value': item_group_total_purchase_df,
                                    })

item_group_summary_sorted_df = item_group_summary_df.sort_values('Purchase Count', ascending=False)
item_group_summary_sorted_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Group by SN
item_group_df = purchase_data.groupby(['Item ID', 'Item Name'])
item_group_count_df = item_group_df['Item ID'].count()
item_group_total_purchase_df = item_group_df['Price'].sum()
item_group_price_df = (item_group_df['Price'].sum()/item_group_df['Price'].count())

item_group_summary_df = pd.DataFrame({
                                    'Purchase Count': item_group_count_df,
                                    'Item Price': item_group_price_df,
                                    'Total Purchase Value': item_group_total_purchase_df,
                                    })

item_group_summary_sorted_df = item_group_summary_df.sort_values('Total Purchase Value', ascending=False)
item_group_summary_sorted_df['Item Price'] = item_group_summary_sorted_df['Item Price'].map('${:.2f}'.format)
item_group_summary_sorted_df['Total Purchase Value'] = item_group_summary_sorted_df['Total Purchase Value'].map('${:.2f}'.format)

item_group_summary_sorted_df.head()
```
